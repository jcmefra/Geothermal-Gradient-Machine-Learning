```python
import pandas as pd

df = pd.read_csv('geo_col_map.csv')
```

```python
def classify_geology(description):
    if not isinstance(description, str):
        return "no"
    
    igneous_keywords = ["basaltos", "ígneas", "granitos", "dioritas", "riolitas", "andesitas", "gabros",
                        "pórfidos", "peridotitas", "tonalitas", "sienitas", "cuarzomonzonitas"]
    metamorphic_keywords = ["esquistos", "gneises", "metamórficas", "cuarcitas", "mármoles", "anfibolitas", "pizarras", "migmatitas",
                            "serpentinitas"]

    description = description.lower()
    if any(keyword in description for keyword in igneous_keywords) or any(keyword in description for keyword in metamorphic_keywords):
        return "yes"
    else:
        return "no"
```

```python
df['Basement'] = df['Descripcion'].apply(classify_geology)
```

```python
df.to_csv('classified_geo_data.csv', index=False)
```
